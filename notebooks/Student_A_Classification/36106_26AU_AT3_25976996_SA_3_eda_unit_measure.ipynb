```python
dataset_insights = """
The unit_measure dataset is a small reference/lookup table with 38 rows and 2 columns (unit_measure_code as UUID, name as text).
It contains no missing values and no duplicates.

This table maps measurement unit codes (e.g., Boxes, Bottle, Centimeters) to their descriptive names.
It serves as a dimension table linked to the product table through size_unit_measure_code and weight_unit_measure_code.

Due to its small size and reference nature, this dataset requires minimal analysis. Its primary use is for enriching product data during feature engineering.
"""
```

### B.3 Explore Feature of Interest — name

```python
print(f"All unit measure names:")
print(df['name'].tolist())
print(f"\nUnique names: {df['name'].nunique()}")

# Categorise by type
df['category'] = df['name'].apply(lambda x: 'Weight' if x in ['Grams','Kilograms','Ounces','Pounds','Milligrams']
                                   else 'Length' if x in ['Centimeters','Meters','Inches','Feet','Millimeters']
                                   else 'Volume' if x in ['Liters','Milliliters','Gallons','Pints','Quarts','Cups','Fluid Ounces']
                                   else 'Packaging' if x in ['Boxes','Bottle','Carton','Case','Dozen','Each','Pack','Pallet']
                                   else 'Other')
print(f"\nUnit categories:\n{df['category'].value_counts()}")
```

```python
feature_2_insights = """
The name column contains 38 unique measurement units spanning multiple categories: weight measures (Grams, Kilograms, etc.), length measures (Centimeters, Inches, etc.),
volume measures (Liters, Gallons, etc.), and packaging units (Boxes, Carton, Dozen, etc.).

This diversity reflects the retailer's varied product range. For the classification use case (customer repeat purchase prediction), unit_measure is unlikely to be a direct feature,
but it enriches product understanding — for example, products sold by weight vs. by unit may correspond to different product categories with different repeat purchase dynamics.
"""
```

```python
# Do not modify this code
print_tile(size="h3", key='feature_2_insights', value=feature_2_insights)
```
